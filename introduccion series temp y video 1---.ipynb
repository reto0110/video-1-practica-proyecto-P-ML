<h3> Remuestreo de datos y cambio de temporalidad de serie </h3>

```python
san.Close.resample('M').ohlc().head() #una forma errornea de calcular las vales de temporalidad Mensual
#el resample hace muestras mensuales, y e ohlc busca la apertura, cierre, mayor y menor de la muestra que le
#resample
```

para ver la temporalidad M verdadera crearemos una util funcion:

```python
def remuestreo(df,temp):
    df.resample(temp).agg({'Open': 'first',
                          'High': 'max',
                          'Low':'min',
                          'Close':'last',
                          'Adj Close':'last',
                          'Volumen':'sum'})
    return df
```

.agg() agrege o aplica funciones a un df sobre sus col (por defecto), por tanto el nuevo df tendra las col (Open, High, Low, etc), cuyo valor vendra dado por el resultado de las funciones (first, max, min, etc) sobre las respectivas cols.

```python
remuestreo(san,'M').head()
```

Ahora s√≠ tenemos el remuestreo y por tanto el cambio de temporalidad de la data a un mes, esto sirve para temp superiores tambien como 3M o Y

```python
def ajustar(df):
    df=df.copy() #hacemos una copia del df input en un df local
    lista=['Open','High','Low']
    for f in lista:
        df[f]=df[f]*df['Adj Close']/df['Close']
    df=df.drop('Close',axis=1) # botamos la col (axis=1 para eliminar col y no row)
    df=df.rename(columns={'Adj Close':'Close'})
    return df
```

Pensado en la eficiencia y escalabilidad definamos una funcion para a√±adir varias EMAs con distintos periodos que se a√±diran con una sola funcion

```python
def add_emas(df, periods, close='Close'):
    #ademas del df recibiremos, una lista con periodos y que seran aplic en precio Close por def.
    import ta
    for period in periods:
        name= f'EMA_{period}'
        df[name]=ta.ema(df[close],period)
    
    return df
```

```python
periodos= [7,14,28,60,90,120,180,200,260,500]
san_emas= add_emas(san,periods=periodos)
san_emas.head()
    
```

```python
emas=[f'EMA_{p}' for p in periodos]
san_emas[emas].plot()
```

<h3> Relaciones entre 2 valores </3>

buscaremos los retornos de las dos acciones

```python
san['rets']=san.Close.pct_change() #cambios procentuales diarios
bbva['rets']=bbva.Close.pct_change()
san.rets.hist(bins=30, alpha=0.7, color='r') #grafico de histograma
bbva.rets.hist(bins=30, alpha=0.7, color='b')# bins es numero de elemnentos a graficar
plt.legend(['SAN','BBVA']) #legendas en el plot
```

pareciera que el SAN tiene mejor retorno. Podemos comparar la similitud entre la misma varible para los dos activos, per ejem, los gaps diarios

```python
san['Gap']=(san.Open - san.Close.shift(1))/san.Close.shift(1)
bbva['Gap']=(bbva.Open - bbva.Close.shift(1))/bbva.Close.shift(1)
san.Gap.hist(bins=30, alpha=0.7, color='r') #grafico de histograma
bbva.Gap.hist(bins=30, alpha=0.7, color='b')
plt.legend(['SAN','BBVA']) #legendas en el plot
```

tiene mayores gaps el SAN. Con los graficos de dispersion podemos observar, por ejemplo, la diferencia entre un gap y la dif entre aprecio de apertura y precio de cierre del mismo d√≠a:

```python
plt.scatter(san.Gap, (san.Close - san.Open)/san.Open)
```

correlacion 0. Tambien la correlaci√≥n es facil de calcular con pandas:

```python
san.Gap.corr((san.Close - san.Open)/san.Open)  #correlacion pract. nula
```

```python
san.rets.corr(bbva.rets) #correlaci√≥n altisima, muy correlacionados
```

```python
san.rets.plot(alpha=0.3) #alpha define la transparencia del color
bbva.rets.plot(alpha=0.2)
plt.legend()
```

Los gaps no estaban relacionados pero al parecer los retornos s√≠. Podemos ver la dif de los retornos para las dos acciones, agregamos la media para comprobra que es cercana a cero. media de 22 periodos.

```python
(san.rets-bbva.rets).plot()
(san.rets-bbva.rets).rolling(22).mean().plot()
```

Posible <i>Pair Trading</i> Ahora veamos la evoluci√≥n de la correlaci√≥n en el tiempo, calulandola para una ventana en desplazamiento de 90 d√≠as. Obaservaremos que no es constante y se ha deriorado desde 2017
